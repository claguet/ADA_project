```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.cross_validation import KFold
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
%matplotlib inline
from scipy import stats
```

# Cluster the political parties based on what they recommend for each votations

For each votations the political parties emmited some recommendations (yes:1, no:-1, freedom of choice:0).


We used these recommandations for the 42 votations available as the feature space allowing an unsupervised clustering of the political parties. 

Before clustering we reduced the dimensionality of the feature space (from 42 to 5) by transforming the data with a PCA and keeping only the 5 first principal components.
K-means algorithm was then applied on the newly defined feature matrix. 

We repeated the process twice: Once with only the political parties that together represent more than 95% of the population's votes for the national council elections of 2015 (these are the parties we kept for further analysis), and a second time with all the parties that emmited some recommendations. 

For this second case we had to exclude 3 parties from the analysis because they only emmited a recommendation in a minority of the cases. For the other partie we assigned a special value to the recommendation when no recommendation was available (2).


```python
# associate oui non liberté de vote to 1, -1, 0
recom={'oui':1, 'non':-1, 'liberté de vote':0}
```

```python
#create political party groups
party=pd.read_excel('Tables/Votation/Partis politiques Suisse.xlsx')
Droite=list(party.Sigle[party.Classification=='Droite'])
Gauche=list(party.Sigle[party.Classification=='Gauche'])
```

```python
english_names={'UDC': 'SVP',
              'PS':'SP',
              'PLR':'FDP',
              'PDC':'CVP',
              'PBD':'BDP',
              'PES':'GPS',
              'PVL':'GLP',
              'PEV':'EVP'}
```

```python
# create a list of the parties we want to keep
good_party=[]
for i in english_names.keys():
    good_party.append(i)
```

```python
good_party
```

```python
# dataframe for the recommendations

# for the parties we want to keep only
Recommandation=pd.DataFrame(index=good_party)

# for all the identified parties
Recommandation_all=pd.DataFrame(index=Droite+Gauche)

# keep the number of all votations
votations=[]
for year in range(2011,2016):
    year=str(year)
    data=pd.read_excel('Tables/Votation/recommandation_parti_votations_federales.xls',sheetname=year,header=6)
    # drop the empty rows
    data.dropna(axis=0, how='any', inplace=True) 
    # remove the 'No ' in the column names
    data.columns=[col[3:] for col in data.columns]
    for col in data.columns:
        Recommandation[col]=0
        Recommandation_all[col]=0
        votations.append(col)
        for parti in good_party:
            Recommandation[col][parti]=recom[data[col][parti]]
        for parti in Droite+Gauche:
            if parti in data.index:
                Recommandation_all[col][parti]=recom[data[col][parti]]
            else: # assign the value 2 to the party that didn't emmit a recommendation
                Recommandation_all[col][parti]=2

```

```python
Recommandation_all
```

```python
#remove the parties that had almost no data (that have a lot of 2)
Recommandation_all.drop(['Lega','DS','MCR'],axis=0, inplace=True)
```

# Reduce the feature space by using PCA

We applied PCA to reduce the dimensions of the recommendation space. We then plotted the data along the two 1st principal components.

The colors in these plots correspond to the labels we had manually assigned to the parties by searching for informations on the wen.

### On the reduced list of parties (the ones we will keep for further analysis)

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=4)
X_new=pca.fit_transform(np.array(Recommandation))
print(pca.explained_variance_ratio_)
```

the 3 1st principal components explain more than 82% of the variance in the data

```python
Gauche_kept=[i for i in Recommandation.index.tolist() if i in Gauche]
```

```python
Droite_kept=[i for i in Recommandation.index.tolist() if i in Droite]
```

```python
# assign colors to each party according to the manual classification previously made
colors=[]
for i in Recommandation.index.tolist():
    if i in Gauche:
        colors.append('red')
    elif i in Droite:
        colors.append('blue')
        
```

```python
x=X_new[:,0]
y=X_new[:,1]
n=Recommandation.index.tolist()

fig, ax = plt.subplots()
ax.scatter(x, y,color=colors)

for i, txt in enumerate(n):
    ax.annotate(txt, (x[i],y[i]))
```

### On all the parties who emmited recommendations

```python
pca = PCA(n_components=5)
X_new_all=pca.fit_transform(np.array(Recommandation_all))
print(pca.explained_variance_ratio_)
```

```python
# assign the colors based on the manually identified groups (left or right)
colors=[]
index_left=[]
index_right=[]
for j,i in enumerate(Recommandation_all.index.tolist()):
    if i in Gauche:
        colors.append('red')
        index_left.append(j)
    elif i in Droite:
        colors.append('blue')
        index_right.append(j)
```

```python
x=X_new_all[:,0]
y=X_new_all[:,1]
n=Recommandation_all.index.tolist()

fig, ax = plt.subplots()
ax.scatter(x, y, color=colors)

for i, txt in enumerate(n):
    ax.annotate(txt, (x[i],y[i]))

plt.xlabel('1st principal component')
plt.ylabel('2nd principal component')
plt.title('Political parties manually clustered')
```

# K-means

With the new feature space we clustered the parties with the K-means algorithm.

The plots display the results of the clustering along the 2 1st principal components. 
The colors represent the labels assigned by the K-means clustering. 

We tried different numbers of clusters, but were mostly interested in 2 clusters, as we wanted to split our parties into left and right categories.

The K-means algortihm based on the recommendations emmited by each parties for 42 votations produced the same binary classification (left/right) than the manual classification we had previously done.


### On the reduced list of parties (the ones we will keep for further analysis)

```python
# now we have 8 points that have 5 features: want to cluster them 

kmeans = KMeans(n_clusters=2, random_state=0).fit(X_new)
kmeans.labels_
```

```python
# assign a color to the parties based on the labels they received from the k-means algorithm
colors=[]
for i in kmeans.labels_:
    if i ==0:
        colors.append('red')
    elif i==1:
        colors.append('blue')
```

```python
x=X_new[:,0]
y=X_new[:,1]
n=Recommandation.index.tolist()

fig, ax = plt.subplots()
ax.scatter(x, y,color=colors)

for i, txt in enumerate(n):
    ax.annotate(txt, (x[i],y[i]))
```

### On all the parties who emmited recommendations

```python
# now we have 8 points that have 5 features: want to cluster them 

kmeans = KMeans(n_clusters=2, random_state=0).fit(X_new_all)
kmeans.labels_
```

```python
colors=[]
for i in kmeans.labels_:
    if i ==0:
        colors.append('blue')
    elif i==1:
        colors.append('red')
    else:
        colors.append('green')
```

```python
x=X_new_all[:,0]
y=X_new_all[:,1]
n=Recommandation_all.index.tolist()

fig, ax = plt.subplots()
ax.scatter(x, y, color=colors)

for i, txt in enumerate(n):
    if txt=='PCS':
        ax.annotate(txt, (x[i],y[i]),(x[i]-0.5,y[i]+0.3))
    elif txt=='PES':
        ax.annotate(txt, (x[i],y[i]),(x[i]+0.2,y[i]+0.1))
    elif txt=='PS':
        ax.annotate(txt, (x[i],y[i]),(x[i]+0.2,y[i]-0.1))
    else: 
        ax.annotate(txt, (x[i],y[i]))

plt.xlabel('1st principal component')
plt.ylabel('2nd principal component')
plt.title('Political parties clustered with K-means: 2 clusters')

fig.savefig('K_means_party_2cluster.eps', format='eps', dpi=1200)
```
