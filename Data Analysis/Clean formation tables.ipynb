```python
formation_2011 = pd.read_excel('Tables/Demographic/formation_2011.xls', index_col = 1)
formation_2012 = pd.read_excel('Tables/Demographic/formation_2012.xls', index_col = 1)
formation_2013 = pd.read_excel('Tables/Demographic/formation_2013.xls', index_col = 1)
formation_2014 = pd.read_excel('Tables/Demographic/formation_2014.xls', index_col = 1)
```

# Clean the data function

```python
def cleanformationdata (data):
    # Header formation
    for i in range(data.shape[1]):
        if pd.isnull(data.iloc[1][i]):
            data.iloc[1][i] = data.iloc[1][i-1]
    col_names = np.array(data.iloc[1]+'\n'+'\n'+data.iloc[2]).tolist()
    data.columns = col_names
        
    # Remove index NaN
    data = data[pd.notnull(data.index)]
    
    # Remove column NaN
    data = data.dropna(axis=1, how='any')
    
    # change index
    data.index.values[0] = 'Swiss'
    
    # remove the *** % and space from the data
    for col in data.columns:
        data[col] = data[col].map(lambda x: str(x).lstrip('*%X ').rstrip('*%X '))
    
    return data
```

```python
formation_2011 = cleanformationdata(formation_2011)
formation_2012 = cleanformationdata(formation_2012)
formation_2013 = cleanformationdata(formation_2013)
formation_2014 = cleanformationdata(formation_2014)
```
