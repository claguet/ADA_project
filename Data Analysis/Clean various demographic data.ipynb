## Population age 

```python
def cleanpopulationdata (data,year):
    # drop the empty rows
    data.dropna(axis=0, how='any', inplace=True)
    
    # fusion the 2 columns names to keep only one and add the year at the end 
    #population_age.columns=['Total_Total_2011',]
    data.columns = [''.join(col).strip() for col in data.columns.values]
    data.columns = [str(col) +'_'+ year for col in data.columns]
    
    # set the index to the good values
    data.reset_index(inplace=True, drop=True)
    data.index=ShortName2
    
    return data
```

```python
#create a dataframe for each year, put in a dictionnary
population_age_all = {}
for year in range(2010,2016):
    year=str(year)
    data=pd.read_excel('Tables/Demographic/population_age_2010-2015.xls',sheetname=year,header=[3,4])
    pop_age=cleanpopulationdata (data,year)
    #save as a csv
    pop_age.to_csv('Tables/Demographic/Clean/population_age_'+year)
    # save in the dictionnary
    population_age_all[year]=pd.DataFrame(pop_age)
```

## permanent residents by various indicators

```python
def cleanpermresdata (data,year,to_drop,new_indice):
    # drop the empty rows
    data.dropna(axis=0, how='any', inplace=True)
    
    # fusion the 2 columns names to keep only one and add the year at the end 
    #population_age.columns=['Total_Total_2011',]
    data.columns = [''.join(col).strip() for col in data.columns.values]
    
    #change the name of the lasts columns because it is not marital status
    columns_names=data.columns.tolist()
    columns_names[15]='Urban core area'
    columns_names[16]='Area influenced by urban cores'
    columns_names[17]='Area beyond urban core influence'
    data.columns=columns_names
    
    data.columns = [str(col) +'_'+ year for col in data.columns]
    
    # drop the things that are not cantons
    data.drop(to_drop, axis=0, inplace=True)
    
    # set the right index

    data.reset_index(inplace=True)
    data['new_index']=new_indice
    data.set_index(data.new_index,drop=True,inplace=True)
    del data.index.name

    data.drop(['index','new_index'],axis=1,inplace=True)

    # remove the - 
    for col in data.columns:
        data[col] = data[col].map(lambda x: str(x).lstrip('-').rstrip('-'))
    
    return data
```

```python
to_drop=['Lake Geneva region','Espace Mittelland','Northwest Switzerland','Eastern Switzerland','Central Switzerland']
new_indice=['Switzerland','VD','VS','GE','BE','FR','SO','NE','JU','BS','BL','AG','ZH','GL','SH','AR','AI','SG','GR','TG','LU','UR','SZ','OW','NW','ZG','TI']
```

```python
year='2015'
data=pd.read_excel('Tables/Demographic/permanent_resident_by_various_indicator.xls',sheetname=year,header=[3,4])
```

```python
#create a dataframe for each year, put in a dictionnary
perm_res_all= {}
for year in range(2010,2016):
    year=str(year)
    data=pd.read_excel('Tables/Demographic/permanent_resident_by_various_indicator.xls',sheetname=year,header=[3,4])
    perm_res=cleanpermresdata (data,year,to_drop,new_indice)
    #save as a csv
    perm_res.to_csv('Tables/Demographic/Clean/permanent_resident_various_indic_'+year)
    # save in the dictionnary
    perm_res_all[year]=pd.DataFrame(perm_res)
```
